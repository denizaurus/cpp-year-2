```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("pastel")
%matplotlib inline

import json
```

```python
with open("results.json", 'r') as json_file:
    data = json.load(json_file)
```

```python
x = np.linspace(1, 100, 100)

plt.figure(figsize=(12,8))
plt.plot(x, data["for_each"], '.-', label='for_each, sequenced')
plt.plot(x, data["for_each_async"], '.-', label='for_each_async')

locs, labels = plt.yticks()
plt.yticks(locs, locs/1000)
plt.ylim(0, 17500)

plt.title("Сравнение производительности for_each и for_each_async", 
          weight='bold', fontsize=15)
plt.xlabel('$10^5$ элементов в контейнере', fontsize=12)
plt.ylabel('Время, затраченное на выполнение алгоритма, мс', fontsize=12)

plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, data["partial_sum"], '.-', label='partial_sum, sequenced')
plt.plot(x, data["inclusive_scan"], '.-', label='inclusive_scan, parallel')

locs, labels = plt.yticks()
plt.yticks(locs, locs/1000)
plt.ylim(0, 25000)

plt.title("Сравнение производительности последовательного partial_sum \
и асинхронного inclusive_scan", fontsize=15)
plt.xlabel('$10^5$ элементов в контейнере', fontsize=12)
plt.ylabel('Время, затраченное на выполнение алгоритма, мс', fontsize=12)

plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, data["inner_product"], '.-', label='inner_product, sequenced')
plt.plot(x, data["transform_reduce"], '.-', label='transform_reduce, parallel')

locs, labels = plt.yticks()
plt.yticks(locs, locs/1000)
#plt.ylim(0, 25000)

plt.title("Сравнение производительности последовательного inner_product \
и асинхронного transform_reduce", fontsize=15)
plt.xlabel('$10^5$ элементов в контейнере', fontsize=12)
plt.ylabel('Время, затраченное на выполнение алгоритма, мс', fontsize=12)

plt.legend(fontsize=12)
plt.grid()
plt.show()
```
